```python
a = np.array([1,2,3,4,5])
a2 = a**2
```

```python
plt.figure(figsize=(10,8))
plt.plot(a,a, label = "x sqaure", color="yellow")
plt.plot(a,a2, label = "x", color="red", marker ="*")
plt.legend()
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("plot graph")
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y1 = np.array([10,13,9,23,5])
y2 = np.array([34,21,67,9,4])
```

```python
plt.bar(x,y1,width=0.3, label="sales in 2018")
plt.bar(x+0.3,y2,width=0.3, color="red", label="sales in 2019")
plt.legend()
plt.show()
```
