```python
### face regonition
```

```python
detector = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
```

```python
img = cv2.imread("image.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # this line will convert image to RGB format form BGR format
                                            # because Matplotlib internally is codded for BGR format
```

```python
plt.imshow(img)
plt.show()
```

```python
faces = detector.detectMultiScale(img, 1.1, 5)  
```

```python
x, y, w, h = faces[1]
```

```python
faces[1]
```

```python
img = cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0), 2 )
```

```python
plt.imshow(img)
```

```python
#plotting a rectangle all faces using a loop
for face in faces:
    x,y,w,h = face
    img=cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
    plt.imshow(img)
```

```python
### WEB CAM DETECTION
```

```python
cam = cv2.VideoCapture(0) #0 stands for front camera
```

```python
while True:
    ret, frame = cam.read()
    
    if ret==False:
        continue
        
    cv2.imshow("frame",frame)
    cv2.waitKey(1)
    
    # for activating key in keyboard
    key_pressed = cv2.waitKey(1) & 0xFF
    
    #now check if key_pressed is 'q'
    if(key_pressed == ord('q')):
        break
    
    
cam.release()
cv2.destroyAllWindows()

    
```

```python
### FACE DETECTION ON WEB CAM
```

```python
while True:
    ret, frame = cam.read()
    
    if ret==False:
        continue
    
    faces = detector.detectMultiScale(frame,1.2,5)
    for face in faces:
        x,y,w,h = face
        frame = cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
        
    cv2.imshow("frame",frame)
    
    
    # for activating key in keyboard
    key_pressed = cv2.waitKey(1) & 0xFF
    
    #now check if key_pressed is 'q'
    if(key_pressed == ord('q')):
        break
    
    
cam.release()
cv2.destroyAllWindows()
```
